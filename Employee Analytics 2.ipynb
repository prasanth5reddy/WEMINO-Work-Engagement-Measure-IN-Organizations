## Lets roll into ML stuff

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

## Let's do Deep Learning

```python
import torch
from torch.utils import data
import torch.nn as nn
import torch.nn.functional as F
import random
```

```python
torch.manual_seed(0)
random.seed(0)
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_X, Y_enc, test_size=0.25, random_state=0)

train_dataloader = data.DataLoader(data.TensorDataset(torch.tensor(x_train, dtype=torch.float), torch.tensor(y_train))
                                  ,batch_size=32, shuffle=True)
test_dataloader = data.DataLoader(data.TensorDataset(torch.tensor(x_test, dtype=torch.float), torch.tensor(y_test)),
                                 batch_size=32, shuffle=False)
len(train_dataloader), len(test_dataloader)
```

```python
class EmpNet(nn.Module):
    def __init__(self):
        super(EmpNet, self).__init__()
        self.fc1 = nn.Sequential(nn.Linear(45, 32), nn.Dropout(0.5))
        self.fc2 = nn.Linear(32, num_of_classes)
        self.relu = nn.ReLU()

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out
```

```python
net = EmpNet()
learning_rate = 0.02
epochs = 100
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate)
```

```python
total_step = len(train_dataloader)
for epoch in range(epochs):
    running_train_loss, running_test_loss = 0, 0
    for i, (data_x, labels) in enumerate(train_dataloader):
        # Forward pass
        outputs = net(data_x)
        loss = criterion(outputs, labels)
        
        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        running_train_loss += loss.item()
        
    with torch.no_grad():
        for data_x in test_dataloader:
            emp_data, labels = data_x
            outputs = net(emp_data)
            loss = criterion(outputs, labels)
            running_test_loss += loss.item()
            
    if (epoch + 1) % 50 == 0:
        print ('Epoch [{}/{}], Step [{}/{}], Train Loss: {:.4f}, Test Loss: {:.4f}' 
               .format(epoch+1, epochs, i+1, total_step, running_train_loss / total_step,
                      running_test_loss / len(test_dataloader)))
```

```python
with torch.no_grad():
    correct = 0
    total = 0
    for data_emp, labels in test_dataloader:
        outputs = net(data_emp)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    print('Test Accuracy: {} %'.format(100 * correct / total))
```
