# Join everything except in and out time

```python
emp_merge_part_df = pd.merge(emp_surv_df, gen_df, on='EmployeeID')
emp_merge_df = pd.merge(emp_merge_part_df, mang_sur_df, on='EmployeeID')
emp_merge_df.info()
```

### Drop rows with null values

```python
emp_merge_df.dropna(inplace=True)
emp_merge_df.info()
emp_merge_df.head()
```

## Lets play with in and out time

```python
in_time_df.info()
time_in_ofc_df = out_time_df.iloc[:, 1:].apply(pd.to_datetime) - in_time_df.iloc[:, 1:].apply(pd.to_datetime)
time_in_ofc_df.head()
```

### Drop holidays

```python
time_in_ofc_df.dropna(axis = 1, how='all', inplace=True)
time_in_ofc_df = time_in_ofc_df.fillna(pd.Timedelta(seconds=0))
for col in time_in_ofc_df.columns:
    time_in_ofc_df[col] = pd.to_timedelta(time_in_ofc_df[col]).astype('timedelta64[m]').astype(int)
time_in_ofc_df.head()
```

```python
time_in_ofc_df['LeavesTaken'] = (time_in_ofc_df == 0).astype(int).sum(axis=1)
time_in_ofc_df
```

```python
time_in_ofc_df['EmployeeID'] = in_time_df['EmployeeID']
time_in_ofc_df = pd.merge(time_in_ofc_df, emp_merge_df[['StandardHours', 'EmployeeID']], on='EmployeeID')
time_in_ofc_df['StandardHours'] = time_in_ofc_df['StandardHours'] * 60

for col in time_in_ofc_df.columns[:249]:
    time_in_ofc_df[col] = time_in_ofc_df[col] - time_in_ofc_df['StandardHours']
time_in_ofc_df.head()
```

```python
time_in_ofc_df
```

```python
time_in_ofc_df['SumRelProd'] = time_in_ofc_df.iloc[:, :-3].sum(axis=1, skipna=True)
time_in_ofc_df
```

```python
time_in_ofc_df['SumRelProd'] = time_in_ofc_df['SumRelProd'] + time_in_ofc_df['LeavesTaken'] * time_in_ofc_df['StandardHours']
```

```python
X = time_in_ofc_df[['EmployeeID', 'LeavesTaken', 'SumRelProd']]
X
```
